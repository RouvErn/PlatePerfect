```python
import cv2
import matplotlib.pyplot as plt

# Path to your video file
video_path = 'camera_A.h264'

# Open the video file
cap = cv2.VideoCapture(video_path)

# Read and display the first 5 frames
num_frames_to_display = 5

for _ in range(num_frames_to_display):
    ret, frame = cap.read()
    if ret:
        # Convert the color of the frame from BGR to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        plt.imshow(frame_rgb)
        plt.axis('off') # To not display axis
        plt.show()
    else:
        break

# Release the video capture object
cap.release()
```
